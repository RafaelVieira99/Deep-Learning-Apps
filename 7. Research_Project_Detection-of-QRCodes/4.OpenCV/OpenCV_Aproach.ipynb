# Using QRCodeDetector


```python
import glob
import cv2
qrDecoder = cv2.QRCodeDetector()
```

```python
def display(im, bbox):
    n = len(bbox)
    for j in range(n):
        cv2.line(im, tuple(bbox[j][0]), tuple(bbox[ (j+1) % n][0]), (255,0,0), 3)
 
    # Display results
    cv2.imshow("Results", im)
```

```python
def detect():
    path = glob.glob(r"..\YOLOv5\dataset todo\*.jpg")
    #path = glob.glob(r"..\..\..\TCDEI\Assignment6\OpenCV\resized\*.jpg")
    nfile=0
    for file in path:
        print("New image :" + file)
        img = cv2.imread(file,1)
        
        decoder = cv2.QRCodeDetector()
        data, points, _ = decoder.detectAndDecode(img)
 
        if points is not None:
            print('Decoded data: ' + data)

            points = points[0]
            for i in range(len(points)):
                pt1 = [int(val) for val in points[i]]
                pt2 = [int(val) for val in points[(i + 1) % 4]]
                cv2.line(img, pt1, pt2, color=(255, 0, 0), thickness=3)

            cv2.imshow('Detected QR code', img)
        

        #cv2.imshow('imag' + str(nfile + 1), img)
        nfile += 1
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    print('aaaa')
```

```python
detect()
```

# Using pyzbar module

```python
import cv2
from pyzbar.pyzbar import decode
 
 
image = cv2.imread('1.jpg')
 
detectedBarcodes = decode(image)
 
for barcode in detectedBarcodes:
     
    (x, y, w, h) = barcode.rect
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 5)
    print(barcode.data)
    print(barcode.type)
    
imS = cv2.resize(image, (960, 540))                # Resize image
cv2.imshow("output", imS)
#cv2.imshow("Image", image)
  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from pyzbar.pyzbar import decode
import numpy as np 
import os
def detectpyzar():
    #ath = glob.glob(r"..\YOLOv5\dataset todo\*.jpg")
    path = glob.glob(r"..\..\..\TCDEI\Assignment6\OpenCV\dataset_todo\*.jpg")
    path2 = glob.glob(r"..\..\..\TCDEI\Assignment6\OpenCV\resultadosPYZBAR")
    nfile=0
    for file in path:
        print("New image :" + file)
        img = cv2.imread(file,1)
        
        detectedBarcodes = decode(img)
        
        for barcode in detectedBarcodes:
            (x, y, w, h) = barcode.rect
            cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 5)
            print(barcode.data.decode('utf-8'))
            print(barcode.type)
            myData = barcode.data.decode('utf-8')
            pts = np.array([barcode.polygon],np.int32)
            pts = pts.reshape((-1,1,2))
            cv2.polylines(img,[pts],True,(255,0,255),5)
            pts2 = barcode.rect
            cv2.putText(img,myData,(pts2[0],pts2[1]),cv2.FONT_HERSHEY_SIMPLEX,
                        0.9,(0,255,0),2)
            imR = cv2.resize(img,(780,1040))
            cv2.imshow('Detected QR code', imR)
        cv2.imwrite('result_' + file, img)

        #cv2.imshow('imag' + str(nfile + 1), img)
        nfile += 1
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    print('aaaa')
```

```python
import glob
import cv2
from pyzbar.pyzbar import decode
import numpy as np 
import os
```

```python
#code=decode(img)
cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)
```

```python
while True:
    success,img = cap.read()
            
    for barcode in decode(img):
        print(barcode.data)
        myData = barcode.data.decode('utf-8')
        print(myData)
        (x, y, w, h) = barcode.rect
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.imshow('Result',img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
